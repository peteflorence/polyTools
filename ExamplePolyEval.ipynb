```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def horner(x, weights):
    coefficients = weights[::-1]
    result = 0
    for i in coefficients:
        result = result * x + i
    return result

weights = [1, 1, 1, 1, 1]

print horner(1, weights)


def plotHorner(w):
    x = np.linspace(0,10,1000)
    y = x * 0.0
    for index,val in enumerate(y):
        y[index] = horner(x[index],w)
    plt.plot(x,y)
    #plt.axis([-3.14, 3.14, 0, 20])
    plt.show()
    
plotHorner(weights)

def evalDerivative(x, weights, derivative):
    val = 0
    for nn in range(derivative, len(weights)):
        product = 1
        for mm in range(0, derivative):
            product = product * (nn - mm)
        val = val + weights[nn] * product * x**(nn - derivative)
    return val


evalDerivative(1, weights, 0)

def plotDerivative(w, derivative):
    x = np.linspace(0,10,1000)
    y = x * 0.0
    for index,val in enumerate(y):
        y[index] = evalDerivative(x[index],w, derivative)
    plt.plot(x,y)
    #plt.axis([-3.14, 3.14, 0, 20])
    plt.show()

for i in range(0, 6):
    plotDerivative(weights, i)
```
